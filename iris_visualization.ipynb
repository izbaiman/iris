```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dff = pd.read_csv('iris.csv')
```

```python
dff.head()
```

```python
dff.describe()
```

```python
dff['Species'] = pd.factorize(dff['Species'])[0]
dff.plot()
```

```python
fig, ax = plt.subplots()
colors = {'iris-setosa' :'r', 'irs-versicolor':'g', 'iris-virginica' :'b'}
ax.scatter(dff['SepalLengthCm'], dff['SepalWidthCm'] , c = dff['Species'])
```

```python
sns.scatterplot(data= dff, x= 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species')
```

```python
sns.set(style = "ticks")
sns.pairplot(dff, hue= "Species", height = 2.5)
plt.show()
```
